```python
target = pd.read_excel(r'C:\Users\Niveatha\OneDrive\Niv\Preppin_Challenges\PD 2024 Wk 3 Input.xlsx',sheet_name= 'Q1')
```

```python
wk1sales['Class'].replace({'Economy': 'First Class', 
                           'First Class': 'Economy', 
                           'Business Class':'Premium Economy', 
                           'Premium Economy':'Business Class'  }, inplace= True)
```

```python
def extractfirstletters (s):
    words = s.split()
    if len(words) ==1:
        return words[0][0]
    else:
        return ''.join(word[0] for word in words)
 
```

```python
wk1sales['Classnew'] =   wk1sales['Class'].apply(extractfirstletters)
wk1sales
```

```python
wk1sales.drop(columns='Class',inplace=True)
wk1sales.rename(columns= {'Classnew': 'Class'},inplace = True)
```

```python
wk1sales['Month'] = wk1sales['Date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d').month)
```

```python
dfsales= wk1sales.groupby(['Class', 'Month'])['Price'].sum()
```

```python
dfsales.reset_index()
```

```python
target
```

```python
result = pd.merge(dfsales, target, on=['Class','Month'], how='inner')
```

```python
result['Difference'] = result['Price']-result['Target']
```
