```python
import pandas as pd
```

```python
df = pd.read_csv('PD 2024 Wk 1 Input.csv')
```

```python
df.head()
```

```python
df[['Date','Flight Number','location','Class','Price']]=df['Flight Details'].str.split('//',expand= True)
df.head()
```

```python
df[['From','To']] = df['location'].str.split('-',expand= True)
```

```python
df.drop(columns=['location', 'Flight Details'])
```

```python
df.rename(columns= {'Flow Card?': 'Flow Card'}, inplace= True)
```

```python
df= df[['Date','Flight Number','From', 'To', 'Class', 'Price', 'Flow Card', 'Bags Checked', 'Meal Type' ]]
```

```python
df.dtypes
```

```python
df['Date']= pd.to_datetime(df['Date'])
df['Price']= df['Price'].astype(float).round(2)
```

```python
df.dtypes
```

```python
df['Flow Card']=df['Flow Card'].replace({1:'Yes', 0:'No'})
df.head()
```

```python
Output1= df[df['Flow Card']=='Yes']
Output2= df[df['Flow Card']=='No']
```
